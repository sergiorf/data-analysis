```python
import pandas as pd
import matplotlib.pyplot as plt

# Load datasets
life_path = "../data/life-expectancy.csv"
vacc_path = "../data/Diphtheria tetanus toxoid and pertussis (DTP) vaccination coverage.csv"

life_df = pd.read_csv(life_path)
vacc_df = pd.read_csv(vacc_path)

# Keep only World
world_life = life_df[life_df["Entity"] == "World"][["Year", "Period life expectancy at birth"]]
world_life.columns = ["Year", "LifeExpectancy"]

global_vacc = vacc_df[vacc_df["GROUP"] == "GLOBAL"][["YEAR", "COVERAGE"]]
global_vacc.columns = ["Year", "Coverage"]

# ---- Plot 1: Life expectancy (long view) ----
plt.figure(figsize=(12,6))
plt.plot(world_life["Year"], world_life["LifeExpectancy"], color="blue")
plt.title("Global Life Expectancy at Birth (17702023)")
plt.xlabel("Year")
plt.ylabel("Life Expectancy (years)")
plt.grid(True)

# Annotate milestones
plt.axvline(1850, color="grey", linestyle="--", alpha=0.7)
plt.text(1850, 35, "Sanitation improvements", rotation=90, va="bottom")

plt.axvline(1945, color="red", linestyle="--", alpha=0.7)
plt.text(1945, 40, "Antibiotics", rotation=90, va="bottom")

plt.axvline(1974, color="green", linestyle="--", alpha=0.7)
plt.text(1974, 50, "WHO Immunization Programme", rotation=90, va="bottom")

plt.show()

# ---- Plot 2: Vaccination coverage ----
plt.figure(figsize=(10,5))
plt.plot(global_vacc["Year"], global_vacc["Coverage"], marker="o", color="green")
plt.title("Global DTP Vaccination Coverage (19802023)")
plt.xlabel("Year")
plt.ylabel("Coverage (%)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data_path = "../data/Diphtheria tetanus toxoid and pertussis (DTP) vaccination coverage.csv"
df = pd.read_csv(data_path)

# Filter only global data
global_df = df[df["GROUP"] == "GLOBAL"]

# Plot coverage over years
plt.figure(figsize=(10,6))
plt.plot(global_df["YEAR"], global_df["COVERAGE"], marker="o", linestyle="-")
plt.title("Global DTP Vaccination Coverage Over Time")
plt.xlabel("Year")
plt.ylabel("Coverage (%)")
plt.grid(True)
plt.show()
```
